```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Jumlah Penduduk Menurut Provinsi di Indonesia, 2024.xlsx", skiprows=2)
df.head()
```

```python
df.columns = ["Provinsi", "Jumlah Penduduk"]
df.head()
```

```python
df["Jumlah Penduduk"] = pd.to_numeric(df["Jumlah Penduduk"])
df.info()
```

```python
print("Info Data:")
print(df.info())
print("\n5 data teratas:")
print(df.head())
```

```python
total_penduduk = df["Jumlah Penduduk"].sum()
print("\nTotal Penduduk Indonesia (2024):", total_penduduk)
```

```python
print("\n5 Provinsi dengan Penduduk Terbanyak:")
print(df.sort_values("Jumlah Penduduk", ascending=False).head(5))
```

```python
print("\n5 Provinsi dengan Penduduk Terkecil:")
print(df.sort_values("Jumlah Penduduk", ascending=True).head(5))
```

```python
plt.figure(figsize=(12,6))
plt.bar(df["Provinsi"], df["Jumlah Penduduk"])
plt.xticks(rotation=90)
plt.title("Jumlah Penduduk per Provinsi di Indonesia (2024)")
plt.ylabel("Jumlah Penduduk")
plt.xlabel("Provinsi")
plt.tight_layout()
plt.show()
```
